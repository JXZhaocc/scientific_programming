```python
f = scipy.interpolate.interp1d(model[:,0],model[:,1])
x = exp[:,0]
y = f(x)
```

```python
def error(f):
    return np.nansum(np.square(np.subtract(exp[11:-2,1],f*y[11:-2])))
```

```python
scipy.optimize.minimize_scalar(error) 
```
