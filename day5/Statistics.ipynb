```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

a. Create a discrete random variable with poissonian distribution and plot its probability mass function (PMF), cummulative distribution function (CDF) and a histogram of 1000 random realizations of the variable

```python
a = stats.poisson(1)
sample = a.rvs(size = 1000)
```

```python
x = np.arange(-10,10)
pmf = a.pmf(x)
plt.plot(x,pmf)
```

```python
cdf = a.cdf(x)
plt.step(x,cdf)
```

```python
plt.hist(sample,bins=10)
```

b. Create a continious random variable with normal distribution and plot its probability mass function (PMF), cummulative distribution function (CDF) and a histogram of 1000 random realizations of the variable

```python
normal = stats.norm(loc=0, scale = 1)
normal_sample = normal.rvs(size = 1000)
```

```python
x_normal = np.arange(-10,10)
pdf_norm = normal.pdf(x)
plt.plot(x_normal,pdf_norm)
```

```python
cdf_norm = normal.cdf(x)
plt.step(x,cdf_norm)
```

```python
plt.hist(normal_sample,bins=10)
```

 c. Test if two sets of (independent) random data comes from the same distribution

```python
stats.ttest_ind(normal_sample, sample)
# since p-value is much smaller than 5%, we reject the h0 hypothesis, and say that the data comes from different distribution
# which is true
```
